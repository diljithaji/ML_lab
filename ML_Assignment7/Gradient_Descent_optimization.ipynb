```python
import numpy as np
import matplotlib.pyplot as plt

def func(val):
    return val * val

def derivative(val):
    return 2 * val

alpha = 0.1         
iterations = 20     

current_x = 3.0     

x_history = []
y_history = []

for step in range(iterations + 1):
    x_history.append(current_x)
    y_history.append(func(current_x))
    
    slope = derivative(current_x)
    current_x -= alpha * slope

x_vals = np.linspace(-4, 4, 400)
y_vals = func(x_vals)

plt.figure(figsize=(6, 3))
plt.plot(x_vals, y_vals, label="y = xÂ²")
plt.scatter(x_history, y_history, label="Descent Path")
plt.plot(x_history, y_history, linestyle="--")
plt.scatter(x_history[-1], y_history[-1], s=80)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Gradient Descent Illustration")
plt.legend()
plt.grid(True)
plt.show()
```
